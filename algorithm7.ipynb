# 삽입정렬

```python
def insert_sort(lst): 
    # 리스트요소 개수–1번 반복 하기 위한 변수  n 
    # key값을 정하는 첫번째 반복문, 1번 위치부터 시작
    for i in range(1, len(lst)): 
        key = lst[i] 
        # key값의 위치를 찾아 삽입하고, 뒤로 한 칸씩 밀어주는 두번째 반복문 
        for j in range(i): 
            if key < lst[j]: 
                lst[i], lst[j] = lst[j], lst[i] 
    return lst
```

```python
import random as r 

# 입력 받은 숫자의 개수만큼 랜덤 수를 가지는 리스트 반환
def make_test_list(n): 
    l = [] 
    # 1부터 100만 사이의 숫자 중 랜덤으로 리스트에 n개만큼 추가 
    for x in range(n): 
        l.append(r.randint(1, 1000000)) 
    return l 

test1 = make_test_list(10000) # 1만 개의 랜덤 수를 가진 리스트
```

```python
insert_sort(test1)
```

```python
import random as r 
import time as t 
test1 = make_test_list(10000) 
start = t.time()  # 정렬을 시작하는 시간 저장 
result = insert_sort(test1) 
end = t.time()  # 정렬이 완료된 시간저장 

# 정렬이 끝난시간에서 시작시간을 빼서 경과된 시간 측정 
print(end - start) 
```
