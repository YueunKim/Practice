# 거품정렬

```python
def bubble_sort(lst): 
    # 리스트 요소 개수 –1 번  반복하기 위한 변수n 
    n = len(lst) - 1 
    # n만큼 반복하는 첫번째 반복문 
    for i in range(n): 
        # 인접 원소를 비교하며 정렬하는 사이클을 위한 반복문 
        # 횟수가 n-i 인 이유는 정렬된 부분을 제외 
        for j in range(n-i): 
            if lst[j] > lst[j+1]: 
                lst[j], lst[j+1] = lst[j+1], lst[j] 
    return lst
```

```python
list = [3,6,2,8,1]
bubble_sort(list)
```

```python
import random as r 

# 입력 받은 숫자의 개수만큼 랜덤 수를 가지는 리스트 반환
def make_test_list(n): 
    l = [] 
    # 1부터 100만 사이의 숫자 중 랜덤으로 리스트에 n개만큼 추가 
    for x in range(n): 
        l.append(r.randint(1, 1000000)) 
    return l 

test1 = make_test_list(10000) # 1만 개의 랜덤 수를 가진 리스트 
```

```python
bubble_sort(test1)
```

```python
import random as r 
import time as t 
test1 = make_test_list(10000) 
start = t.time()  # 정렬을 시작하는 시간 저장 
result = bubble_sort(test1) 
end = t.time()  # 정렬이 완료된 시간저장 

# 정렬이 끝난시간에서 시작시간을 빼서 경과된 시간 측정 
print(end - start) 
```
